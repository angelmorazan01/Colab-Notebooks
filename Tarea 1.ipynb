__Ejercicio 4:__ Busca tres ejemplos de ecuaciones generales de circunferencia, parábola y elipse fuera del origen y usando la función `plot_implicit` grafíquelas. 

Nota: Dé valores a las constantes $A,B,C,D,E,F \in \mathbb{R}$.

1) Círculo:  $x^{2}+y^{2}+Cx+Dy+F=0$ 

*Función:* $x^2 + y^2 - 2x +2y - 1 = 0$

```
from sympy import plot_implicit, Eq
plot_implicit(Eq(x**2 + y**2 - 2*x +2*y - 1, 0), (x,-2, 3), (y, -3, 2))
```

2) Parábola: $y^{2}+Cx+Dy+F=0$

*Función:* $x^2 -6x - 8y - 7 = 0$

Gráfica: 


```
plot_implicit(Eq(x**2 -6*x - 8*y - 7, 0), (x,-5, 10), (y, -2, 4))
```

3) Elipse: $Ax^{2}+By^{2}+Cx+Dy+F=0$ 

*Función:* $x^2 + 4y^2 - 6x - 16y - 21 = 0$

Gráfica:

```
plot_implicit(Eq(x**2 + 4*y**2 - 6*x - 16*y - 21, 0), (x,-5, 10), (y, -2, 6))
```

**Ejercicio 9:** Escoja tres de las siguientes funciones, grafíquelas fijando de manera adecuada el correspondiente dominio (intervalo) y agregue un leyenda al plano que indique la gráfica que corresponde a cada función. Ajuste de manera adecuada los tamaños de paso para que NO sobresalga una función más que las restantes. Use la librería `matplotlib`.

(a) $y = \frac{1}{x^2}$

(b) $y = tan(x)$

(c) $y = \sqrt{-x}$

(d) $y = ln(x)$

(e) $y = e^x$

(f) $y = 2^x$


Funciones a graficar:\
$y_1= tan (x)$ \
$y_2= \sqrt{-x}$ \
$y_3=e^x$

```
x1 = np.arange (-0.49*np.pi, 0.49*np.pi, 0.01)
x2 = np.arange (-4,0,0.01)
x3 = np.arange (0,3.5,0.01)

y1 = np.tan (x1)
y2 = np.sqrt (-x2)
y3 = np.exp (x3)

mplot.plot (x1, y1, label='$y1 = tan (x)$')
mplot.plot (x2,y2, label= '$y2 = \sqrt {-x}$')
mplot.plot (x3,y3, label='$y3 = e^x$')
mplot.legend()
mplot.show
```
