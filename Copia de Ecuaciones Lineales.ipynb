**Ejercicio 1:** Determine la intersección de las rectas siguientes:
$$-x-2y+3=0$$
$$3x+2y-1=0$$




```
import matplotlib.pyplot as mplot
import numpy as np

A= np.matrix([[-1, -2], [3, 2]]) # declaramos la matriz A
b = np.matrix([[-3], [1]]) # declaramos el vector b

print(np.linalg.solve(A, b)) # resolvemos el sistema Ax=b

x=np.arange(-2,2,0.01)
y1=-(1.0/2)*x+(3.0/2)
y2=-(3.0/2)*x+(1.0/2)

mplot.plot(x, y1,label="Ecuacion 1")
mplot.plot(x, y2, label = "Ecuacion 2")
mplot.legend ()
mplot.show
```
