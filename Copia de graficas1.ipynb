# Importando matplotlib

<br>

<font size="4">
    
- Usualmente se agrega la instrucción $\texttt{import matplotlib.pyplot as plt}$
<br><br>        
- Con esta instrucción todas las funciones se manejan con el prefijo $\texttt{plt.}$

</font>

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16]);
```

# En toda gráfica hay dos tipos de elementos principales

<br>

<font size="4">
    
- Los datos por graficar, en el caso de la gráfica del ejemplo se tienen dos listas que corresponden al eje de las **abcisas**: $\texttt{[1, 2, 3, 4]}$. El eje de las **ordenadas**: $\texttt{[1, 4, 9, 16]}$.
<br><br>        
- Los elementos *decorativos / informativos* de cada gráfica: etiquetas, colores, divisiones, fuentes, textos, etc.  

</font>

# Ejemplo 1 - graficando datos
 

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16]);
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro');
plt.grid(True)
```

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], [1, 2, 3, 4], [1, 4, 9, 16], 'ro');
plt.grid(True)
```

# Ejemplo 1 - graficando funciones
 
<br>

<font size="4">
    
- Si se conoce una expresión algebraica generalmente utilizan las funciones numpy.arange función $\texttt{numpy.linspace}$ para generar los puntos del eje de las abcisas en que se graficará la función.

</font>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 5, 0.2)

plt.plot(x, x, 'r--', x, x**2, 'bs', x, x**3)
plt.show()

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], [1, 2, 3, 4], [1, 4, 9, 16],'ro')
plt.grid(True)

print(x)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, num=30)

plt.plot(x, x, 'r--', x, x**2, 'bs', x, x**3, 'g^')
plt.show()

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], [1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.grid(True)

print(x)
```
