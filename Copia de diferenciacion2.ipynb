
#### Calculando la derivada con Scipy

https://docs.scipy.org/doc/scipy/reference/generated/scipy.misc.derivative.html

```python
import numpy as np
from scipy.misc import derivative

# El parámetro de la funcion se puede pasar como python-función 
def f(x):
    return x**3 + x**2

# El parámetro de la funcion se puede pasar como definición lambda
g = lambda x: x**3 + x**2

print(derivative(f, 1.0, dx=1e-6))
print(derivative(g, 1.0, dx=1e-6))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative

def ff(x):
    return 3 * np.exp(x) / (x**2 + x + 1)

x = np.linspace(-3,3,100)
f = lambda x: 3 * np.exp(x) / (x**2 + x + 1)
y = ff(x)

a0 = f(0)
a1 = derivative(ff, 0, dx=0.001, n = 1)
a2 = derivative(ff, 0, dx=0.001, n = 2) / 2
a3 = derivative(ff, 0, dx=0.001, n = 3, order=5) / 6

print(a0, a1, a2, a3)

T3 = a0 + a1*x + a2*x**2 + a3*x**3

plt.plot(x, y)
plt.plot(x,T3)
plt.xlim([-3,3])
plt.ylim([0,5])
plt.grid(True)
```
