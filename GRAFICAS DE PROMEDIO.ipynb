```
#Longitud 1
import matplotlib.pyplot as mplot
import numpy as np

x= np.arange(0,1.5,0.01)
y= 0.2748*np.sin(9.190*x +4.484)

mplot.plot(x,y)
mplot.xlabel("t(s)")
mplot.ylabel("θ (Radianes)")
mplot.show ()
```
