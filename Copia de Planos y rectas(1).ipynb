# Planos y Rectas

**Ejercicio 1:** Determine todas las formas de la ecuación de la recta que pasa por los puntos $A=(1,3)$ y $B=(2,3)$. Grafique la ecuación de la recta usando `Matplotlib` o `Sympy`.

```
import matplotlib.pyplot as mplot
import numpy as np

x=np.arange(-1,1,0.01)
y=0*x+3
mplot.plot(x, y)
mplot.show
```

**Ejercicio 2:** Determine todas las formas de la ecuación de la recta con dirección $(1,1,1)$ y apoyada en el punto $(0,0,-1)$. Grafique la ecuación de la recta usando `Matplotlib` o `Sympy`.

```
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax= fig.gca (projection="3d")

t3= np.linspace(-5,5,50)
x=t3
y=t3
z=t3-1

ax.plot(x,y,z)
plt.show()
```
