```python
import matplotlib.pyplot as plt

plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gr√°fica')

plt.plot([1, 2, 3, 4], [1, 4, 9, 16]);
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 50)
plt.plot(x, x**2, '*', label='x cuadrada')
plt.plot(x, x**3, '--', label='x c√∫bica')
plt.plot(x, x**4, '+')
plt.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 10, num=100, endpoint=True)
s = np.cos(4 * np.pi * t)**2 + 2

plt.plot(t, s)
plt.xlabel('$tiempo (s)$')
plt.ylabel('$(4\pi t)^2 + 2$')
plt.title('$\sum_{x=0}^{10}x^2$')

plt.savefig('grafica.png', dpi=300)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(2,3,4)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


x = np.linspace(-4*np.pi,4*np.pi,50)
y = np.linspace(-4*np.pi,4*np.pi,50)
z = x**2 + y**2

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x,y,z)
```
